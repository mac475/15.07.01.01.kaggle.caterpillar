#1. component 계열의 dataset 전체통합

- component dataset의 merge 필요성

    → tube → bill에서 발견되는 component 정보는 type별로 adaptor~threaded까지 다양한 종류별로 특성화된 feature를 보유
    → 확보하여 분석목적
    
- 단, 각 component dataset을 보완필요 있는지 사전확인후, 적절한 작업의 병행필요

-<font color='red'><b>본 dataset merge는 meta dataset들간의 통합임</b></font>

```python
import pandas as pd
import numpy as np
import os

base_location = './dataset/01.original.dataset/'
df_components = pd.read_csv( './dataset/02.ml.verified.dataset/components.verified.csv' )
is_first = True

suffix = 0
for root, dirs, files in os.walk( base_location ) :
    for file in files :
        print( file )
        if file.startswith( 'comp_' ) :
            df = pd.read_csv( root + file )
            
            if is_first == True : 
                df_comp_merged = df_components.merge( df, how = 'left', on = 'component_id' )
                is_first = False
            else :
                df_comp_merged = df_comp_merged.merge( df, how = 'left', on = 'component_id' )
                
            suffix += 1

df_comp_merged.head( 10 )
```
