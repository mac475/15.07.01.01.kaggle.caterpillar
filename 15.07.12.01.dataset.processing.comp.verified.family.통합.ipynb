```python
import pandas as pd
import numpy as np
import os

base_location = './dataset/comp.verified/'
df_components = pd.read_csv( './dataset/02.ml.verified.dataset/components.verified.csv' )
is_first = True
# df_comp_merged = df_components

suffix = 0
for root, dirs, files in os.walk( base_location ) :
    for file in files :
        if file.startswith( 'comp_' ) :
            print( file )            
            df = pd.read_csv( root + file )
            
            if is_first == True : 
                df_comp_merged = df_components.merge( df, how = 'left', on = 'component_id' )
                is_first = False
            else :
                df_comp_merged = df_comp_merged.merge( df, how = 'left', on = 'component_id' )
                
            suffix += 1

df_comp_merged.head( 10 )
```
